```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
label_encoder = LabelEncoder()
```

```python
df = datasets.copy()
```

```python
destinasi = df['destinasi_wisata'] 
jumlah = df['Jumlah']

data = pd.concat([df['destinasi_wisata'], df['Jumlah']], axis=1)
```

```python
data['destinasi'] = label_encoder.fit_transform(data['destinasi_wisata'])

X = data[['Jumlah','destinasi']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia_values = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X[['Jumlah', 'destinasi']])
    inertia_values.append(kmeans.inertia_)
    
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster Optimal')
plt.show()
```

```python
from scipy.spatial.distance import cdist
from math import radians, sin, cos, sqrt, atan2
k = 5
```

```python
data_rad = np.radians(X_scaled)
```

```python
def haversine(coord1, coord2):
    # Radius bumi dalam kilometer
    R = 6371.0

    lat1, lon1 = coord1
    lat2, lon2 = coord2

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance
```

```python
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
kmeans.fit(data_rad)
```

```python
# Mendapatkan pusat cluster dalam radian
cluster_centers_rad = kmeans.cluster_centers_
```

```python
# Menghitung jarak Haversine antara titik data dan pusat cluster
distances = np.array([[haversine(data_point, center) for center in cluster_centers_rad] for data_point in data_rad])
```

```python
cluster_labels = np.argmin(distances, axis=1)
```

```python
# Visualisasi hasil clustering
cmap = plt.cm.viridis
norm = plt.Normalize(vmin=min(cluster_labels), vmax=max(cluster_labels))
scatter = plt.scatter(X_scaled[:, 1], X_scaled[:, 0], c=cluster_labels, cmap=cmap, norm=norm)

# Menambahkan pusat cluster
plt.scatter(cluster_centers_rad[:, 1], cluster_centers_rad[:, 0], s=200, c='red', label='Centroids')
# Menambahkan legenda
plt.legend(handles=scatter.legend_elements()[0], title="Clusters")

# Menambahkan label sumbu
plt.xlabel("Jumlah")
plt.ylabel("Destinasi")

# Menampilkan plot
plt.show()

# Output label cluster
print("Hasil Clustering K-Means:")
print(cluster_labels)
```
